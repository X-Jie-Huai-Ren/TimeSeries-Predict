```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
from math import sqrt

# 设置字体为SimHei（黑体）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题
```

```python
# cnn model
ea=event_accumulator.EventAccumulator(r'./logs/[06-04]16.59.57/cnn/events.out.tfevents.1717491597.lemonlover.16456.0') 
ea.Reload()
cnn_train_rmse = [sqrt(i.value) for i in ea.scalars.Items('Train/loss')]
cnn_train_r2 = [i.value for i in ea.scalars.Items('Train/r2')]
cnn_eval_rmse = [sqrt(i.value) for i in ea.scalars.Items('Eval/loss')]
cnn_eval_r2 = [i.value for i in ea.scalars.Items('Eval/r2')]
```

```python
# lstm model
ea=event_accumulator.EventAccumulator(r'logs\[06-04]17.18.13\lstm\events.out.tfevents.1717492693.lemonlover.20496.0') 
ea.Reload()
lstm_train_rmse = [sqrt(i.value) for i in ea.scalars.Items('Train/loss')]
lstm_train_r2 = [i.value for i in ea.scalars.Items('Train/r2')]
lstm_eval_rmse = [sqrt(i.value) for i in ea.scalars.Items('Eval/loss')]
lstm_eval_r2 = [i.value for i in ea.scalars.Items('Eval/r2')]
```

```python
# lstm model
ea=event_accumulator.EventAccumulator(r'logs\[06-06]13.16.37\seq2seq\events.out.tfevents.1717650997.lemonlover.15996.0') 
ea.Reload()
seq2seq_train_rmse = [sqrt(i.value) for i in ea.scalars.Items('Train/loss')]
seq2seq_train_r2 = [i.value for i in ea.scalars.Items('Train/r2')]
seq2seq_eval_rmse = [sqrt(i.value) for i in ea.scalars.Items('Eval/loss')]
seq2seq_eval_r2 = [i.value for i in ea.scalars.Items('Eval/r2')]
```

```python
# 训练集rmse损失下降图
plt.figure(figsize=(6, 4))
x = range(len(cnn_train_rmse))
plt.plot(x, cnn_train_rmse, color='r', label='cnn')
plt.plot(x, lstm_train_rmse, color='g', label='lstm')
plt.plot(x, seq2seq_train_rmse, color='b', label='seq2seq')
plt.title("训练集rmse损失下降图")
plt.legend()
plt.show()
```

```python
# 测试集rmse损失下降图
plt.figure(figsize=(6, 4))
x = range(len(cnn_eval_rmse))
plt.plot(x, cnn_eval_rmse, color='r', label='cnn')
plt.plot(x, lstm_eval_rmse, color='g', label='lstm')
plt.plot(x, seq2seq_eval_rmse, color='b', label='seq2seq')
plt.title("测试集rmse损失下降图")
plt.legend()
plt.show()
```

```python
# 训练集r2-score迭代图
plt.figure(figsize=(6, 4))
x = range(len(cnn_train_r2))
plt.plot(x, cnn_train_r2, color='r', label='cnn')
plt.plot(x, lstm_train_r2, color='g', label='lstm')
plt.plot(x, seq2seq_train_r2, color='b', label='seq2seq')
plt.title("训练集r2-score迭代图")
plt.legend()
plt.show()
```

```python
# 测试集r2-score迭代图
plt.figure(figsize=(6, 4))
x = range(len(cnn_eval_r2))
plt.plot(x, cnn_eval_r2, color='r', label='cnn')
plt.plot(x, lstm_eval_r2, color='g', label='lstm')
plt.plot(x, seq2seq_eval_r2, color='b', label='seq2seq')
plt.title("测试集r2-score迭代图")
plt.legend()
plt.show()
```
